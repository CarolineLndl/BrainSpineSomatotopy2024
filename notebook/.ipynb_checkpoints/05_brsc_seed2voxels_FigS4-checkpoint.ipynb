## <font color=#B2D732> <span style="background-color: #4424D6"> B/ Data extraction </font></span>
### <font color=#4424D6> I. Mean correlation value extraction </font>


```python
# value extraction
from nilearn.maskers import NiftiMasker
from nilearn import image
import numpy as np
masker=[]
corr_value={};corr_mean={};indiv_f={}
mask_f=[]
WTA_mask_dir=config["second_level"] + "/WinnerTakeAll/gm/"  ## add in the config file
seed_mask=["C1","C2","C3","C4","C5","C6","C7"]
for mask_nb,mask_name in enumerate(seed_mask):
    mask_f.append(glob.glob(WTA_mask_dir + "/WTA_zcorr_thr0_cluster100_s0SMC_"+mask_name+"_cluster50.nii.gz")[0])
    masker.append(NiftiMasker(mask_f[mask_nb],smoothing_fwhm=None, t_r=1.55,low_pass=None, high_pass=None)) # seed masker
    corr_value[mask_name]={};corr_mean[mask_name]={}
    
    for seed_nb, seed_name in enumerate(config["seeds"]["seed_names"]):
        mean_group_f[seed_name]=glob.glob(config["first_level"] + seed_name + "/"+config["targeted_voxels"]["target_name"] + "_fc_maps/Corr/bi-corr_" + str(len(config["list_subjects"])) + "*mean.nii.gz")[0]
        indiv_f[seed_name]=[]
        corr_value[mask_name][seed_name]={}
        corr_mean[mask_name][seed_name]={}
        for ID_nb, ID_name in enumerate(config["list_subjects"]):
            indiv_f[seed_name].append(glob.glob(config["first_level"] + seed_name + "/"+config["targeted_voxels"]["target_name"] + "_fc_maps/Corr/bi-corr_sub-" + ID_name+ ".nii.gz")[0])
            corr_value[mask_name][seed_name][ID_name]=masker[mask_nb].fit_transform(indiv_f[seed_name][ID_nb]) #low_pass=0.1,high_pass=0.01
            corr_mean[mask_name][seed_name][ID_name]=np.nanmean(corr_value[mask_name][seed_name][ID_name],axis=1) # mean time serie

```

```python
# Create a list to store the data in the desired format
data_list = []

# Iterate through the dictionary and extract the data
for outer_key, inner_dict in corr_mean.items():
    for inner_key, subject_dict in inner_dict.items():
        for subject, value in subject_dict.items():
            data_list.append({
                'seeds': outer_key,
                'masks': inner_key,
                'IDs': subject,
                'corr': value[0]
            })

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data_list)

# Display the DataFrame
print(df)
```

```python
# value extraction
from nilearn.maskers import NiftiMasker
from nilearn import image
import numpy as np
masker=[]
corr_value={};corr_mean={}
mask_f=[]
WTA_mask_dir=config["second_level"] + "/WinnerTakeAll/gm/"  ## add in the config file
seed_mask=["C1","C2","C3","C4","C5","C6","C7"]
for mask_nb,mask_name in enumerate(seed_mask):
    mask_f.append(glob.glob(WTA_mask_dir + "/WTA_zcorr_thr0_cluster100_s0SMC_"+mask_name+"_cluster50.nii.gz")[0])
    masker.append(NiftiMasker(mask_f[mask_nb],smoothing_fwhm=None, t_r=1.55,low_pass=None, high_pass=None)) # seed masker
    corr_value[mask_name]={};corr_mean[mask_name]={}
    
    for seed_nb, seed_name in enumerate(config["seeds"]["seed_names"]):
        mean_group_f[seed_name]=glob.glob(config["first_level"] + seed_name + "/"+config["targeted_voxels"]["target_name"] + "_fc_maps/Corr/bi-corr_" + str(len(config["list_subjects"])) + "*mean.nii.gz")[0]
        corr_value[mask_name][seed_name]=masker[mask_nb].fit_transform(mean_group_f[seed_name]) #low_pass=0.1,high_pass=0.01
        corr_mean[mask_name][seed_name]=np.nanmean(corr_value[mask_name][seed_name],axis=1) # mean time serie
```

```python
# plotting
data=corr_mean
for outer_key in data:
    for inner_key in data[outer_key]:
        data[outer_key][inner_key] = data[outer_key][inner_key]

# Create a list of tuples containing the outer key, inner key, and value
data_list = [(outer_key, inner_key, value) 
             for outer_key, inner_dict in data.items() 
             for inner_key, value in inner_dict.items()]

# Convert the list of tuples into a DataFrame
df = pd.DataFrame(data_list, columns=['seed', 'mask', 'corr value'])

# Pivot the dataframe to get a matrix format
heatmap_data = df.pivot_table(index='seed', columns='mask', values='corr value')

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', cbar=True)
plt.title('Heatmap of DataFrame')
plt.show()
```
